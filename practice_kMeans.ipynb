```python
"""
本程序为以《机器学习实战》第10章的代码为基础，但是整体代码完全出自作者之手，而非在原有代码基础上修改
程序的基本架构：
由于k-means算法本身的特点，因此没有设置__init__()方法
本程序以MNIST的训练集的全部60000个样本作为测试集
测试结果错误率大约在41%
实验中初始聚类中心centroids选择为10个数字各在样本集中取一幅图

_setup方法负责初始化参数：包括聚类中心self.centroids等
接受的数据类型为：设计矩阵， 类标向量（非one-hot)

算法架构采用同时维护一个向量和一个二维列表以及聚类中心集
self.current_labels #标记每个样本当前类聚类类标
self.index_clusters #样本索引聚类簇列表
self.centroids #聚类中心坐标集
迭代更新样本聚类结果和聚类中心坐标，直到收敛
总体看速度是该算法的优势，但精度不佳，至少对于MNIST数据集而言

k-means算法的特点：
1.不需要训练集，或者说测试集本身就是训练集；
2.一般来说应该设置一个验证集用来调整聚类数k和选择距离函数等；
3.k-means算法计算速度快，其聚类时间甚至快于逻辑回归，虽然后者作为一个分类算法并不与前者具有可比性，算法的时间复杂度为O(NKt)，N是样本数，K是聚类数，t是迭代次数
4.k-means和kNN一样，是一种通过计算数据点间距离进行聚类/分类的算法，因此：
a.)性能都依赖于距离函数的选择
b.)对样本特征的区分度要求高，换而言之，依赖于特征选择与数据预处理
虽然PCA和ZCA都没有表现出对k-means算法性能的明显帮助，但是可以想象的到大量低区分度的特征会严重干扰距离函数的计算结果
这也是k-means, knn等算法与神经网络模型的典型区别，前两者更像一种算法而后这则多少有一些“智能"
就好象前两者只能理解中文，一旦出现英文表达的数据就束手无策，而后者能够通过训练来在一定程度上学会理解英语，继而提高数据处理的性能表现
"""
import numpy
import theano
import theano.tensor as T
from itertools import izip
import time
import pylab
from scipy.spatial.distance import cdist

def euclid_dist(x, y):
    
    return numpy.sqrt(numpy.sum((x - y)**2, axis=1))

class KMeans(object):
    
    def _setup(self, k, num_features, dataset, labels=None, irange=None, centroids=None):
        """
        如果没有指定中心点，则用均匀分布初始化若干个中心点
        绑定数据集
        """
        self.k = k
        self.numpy_rng = numpy.random.RandomState(seed=19920130)
        if centroids is not None:
            assert irange is None
            assert isinstance(centroids, numpy.ndarray)
            assert centroids.shape[1] == num_features
        else:
            #初始化中心点, 该处初始化方法也是用于其他有标数据集
            #assert irange is not None
            assert labels is not None
            mark = [None for i in xrange(self.k)] 
            record = range(self.k)
            cnt = self.k
            centroids = []
            for i in xrange(dataset.shape[0]):
                if mark[labels[i]] == None:
                    mark[labels[i]] = 1
                    record[self.k - cnt] = labels[i]
                    cnt -= 1
                    centroids.append(dataset[i])
                    if cnt == 0:
                        break
                        
            assert len(centroids) == self.k
            centroids = numpy.array(centroids)
        self.centroids_record = record        
        self.centroids = centroids #聚类中心坐标字典，是一个ndarray
        assert isinstance(dataset, numpy.ndarray)
        assert dataset.ndim ==  2
        self.dataset = dataset
        self.features = num_features
        self.current_labels = numpy.zeros(shape=dataset.shape[0]) #标记每个样本当前类聚类类标
        batch_size = self.dataset.shape[0] / self.k
        self.index_clusters = [[] for i in xrange(self.k)] #样本索引聚类簇列表
        for i in xrange(self.k):
            self.current_labels[i*batch_size:(i+1)*batch_size] = i
            self.index_clusters[i].extend(xrange(i*batch_size, (i+1)*batch_size))
        assert isinstance(labels, numpy.ndarray)
        assert labels.ndim == 1
        self.labels = labels
        
    def plot_centroids(self, rows, cols):
        
        for i in xrange(rows):
            or j in xrange(cols):
            pylab.subplot(rows, cols, i*cols+j+1); pylab.axis('off'); pylab.imshow(self.centroids[i*cols+j].reshape(28, 28), cmap=cm.gray)
        pylab.show()
        
        
    def update_class(self):
        """更新样本聚类结果"""
        change_mark = False
        #index_clusters = [[] for i in xrange(self.k)]
        #l_index = [[] for i in xrange(self.dataset.shape[0])]
        #考虑改成mini_batch代码
        for index, sample in enumerate(self.dataset):
            dist_vector = euclid_dist(sample, self.centroids) # float64
            class_num = numpy.argmin(dist_vector)
            if class_num != self.current_labels[index]: #是否被修改
                change_mark = True
                self.index_clusters[int(self.current_labels[index])].remove(index) #从旧的索引簇中删除当前样本索引
                self.index_clusters[class_num].append(index) #在新的簇中加入当前样本索引
                self.current_labels[index] = class_num #修改分类记录，记录当前样本被分到的新类标
            #self.l[class_num].append(sample)
            #l_index[class_num].append(self.index[index])
            #self.index = numpy.array(l_index) #更新索引列表
            
        return change_mark
    
    def update_centroids(self):
        """更新聚类中心"""
        for i in xrange(self.k):
            self.centroids[i] = (self.dataset[self.index_clusters[i]]).mean(axis=0)
        
    def train(self, dataset, labels, num_epoch, k, num_features, irange=None, centroids=None):
        """k_means算法主体"""
        self._setup(k=k, num_features=num_features, dataset=dataset, labels=labels, irange=irange, centroids=centroids)
        print 'setup finish ...'
        
        time_list = []
        for i in xrange(num_epoch):
            print 'epoch', i
            #for i in xrange(2):
            #    for j in xrange(5):
            #        pylab.subplot(2, 5, i*5+j+1); pylab.axis('off'); pylab.imshow(self.centroids[i*5+j].reshape(28, 28), cmap=cm.gray)
            #pylab.show()
            time_list.append(time.clock())
            continue_learning = self.update_class()
            if continue_learning is False:
                break
            time_list.append(time.clock())
            #print '    time elapse on update class:', time_list[-1] - time_list[-2]
            self.update_centroids()
            time_list.append(time.clock())
            #print '    time elapse on update centroids:', time_list[-1] - time_list[-2]
            print '    time elapsed on epoch %d is %f' % (i+1, time_list[-1] - time_list[-3])
            #print [len(i) for i in self.index_clusters]
            print 'initial:', self.error_rate(mode='initial')
            print 'major:', self.error_rate(mode='major')
            print 'nearest:', self.error_rate(mode='nearest')
        print 'total time is', time_list[-1] - time_list[0]
        #print 'initial:', self.error_rate(mode='initial')
        #print 'major:', self.error_rate(mode='major')
        #print 'nearest:', self.error_rate(mode='nearest')
        
    def get_predict(self, mode='initial'): #initial参数用来判定是否使用初始点的类标来作为最终聚类中心的类标
        """获得预测类标"""
        assert mode in ['initial', 'major', 'nearest']
        if mode == 'initial':
            predict = numpy.zeros(shape=(self.dataset.shape[0]))
            for k, cluster in enumerate(self.index_clusters):
                predict[cluster] = self.centroids_record[k]
        elif mode == 'major':
            predict = numpy.zeros(shape=(self.dataset.shape[0]))
            for k, cluster in enumerate(self.index_clusters):
                l = numpy.zeros(10)
                for i in cluster:
                    #print 'cluster:',cluster
                    l[self.labels[i]] += 1
                predict[cluster] = numpy.argmax(l)
        else:
            predict = numpy.zeros(shape=(self.dataset.shape[0]))
            label_dict = numpy.zeros(self.k)
            for k, index_cluster in enumerate(self.index_clusters):
                sample_cluster = self.dataset[index_cluster]
                dists = cdist([self.centroids[k]], sample_cluster) #距离索引为k的聚类中心最近的该类中的点
                #print 'dists.shape:', dists.shape
                closest = numpy.argmin(dists) #closest是一个向量
                label_dict[k] = self.labels[int(index_cluster[closest])]
            for k, cluster in enumerate(self.index_clusters):
                predict[cluster] = label_dict[k]
                
        return predict
        
    def error_rate(self, mode='initial'):
        """计算分类错误数和错误率"""
        predict = self.get_predict(mode=mode)
        error_num = numpy.sum(predict != self.labels)
        error_rate = numpy.cast['float64'](error_num) / self.dataset.shape[0]
        
        return [error_num, error_rate]
        
if __name__ == '__main__':
    
    #import os
    #print os.getcwd()
    from pylearn2.datasets.mnist import MNIST
    #from pylearn2.datasets.preprocessing import ZCA
    from preprocessing import PCA
    
    #preprocessor = ZCA()
    #preprocessor = PCA(num_components=50)
    dsm_train = MNIST(which_set='train', start=0, stop=60000)
    #preprocessor.apply(dsm_train, can_fit=True)
    #print 'features=', dsm_train.X.shape[1]
    k_means = KMeans()
    k_means.train(dataset=dsm_train.X, labels=dsm_train.y, num_epoch=100, k=10, num_features=784, irange=1., centroids=None)
    
```
