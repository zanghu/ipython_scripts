```python
"""
当前cell内的代码已经被废弃，正确的代码见下一个cell
错误原因是在对数据进行预处理的函数preprocess()有隐含的逻辑错误
即其在处理测试数据时，减去了测试样本集的均值并除以测试样本集的标准差，这是错误的
正确的做法是测试集也要减去训练集的均值并除以训练集的标准差。
保留概算法的目的在于对比正确和错误两种方法的计算结果差异
"""
import numpy
import theano
import theano.tensor as T
import os
import operator

def preprocess(X):
    
    assert isinstance(X, numpy.ndarray) and X.ndim == 2
    feature_mean = X.mean(axis=0)
    std = X.std(axis=0)
    
    return (X - feature_mean) / std

class KNN_deprecated(object):
    
    def __init__(self, k, X, y, is_preprocess=True):

        assert isinstance(X, numpy.ndarray) and X.ndim == 2
        assert isinstance(y, numpy.ndarray) and y.ndim == 1
        self.k = k
        self.is_preprocess = is_preprocess
        if self.is_preprocess is True:
            X = preprocess(X)
        self.X = X
        self.y = y
        self.num_sample = y.shape[0]
        #print 'num_samples:', self.num_sample  
   
    def classify(self, X, y=None):
        
        assert isinstance(X, numpy.ndarray)
        if X.ndim == 1:
            X = numpy.array([X], dtype=X.dtype)
        assert X.ndim == 2
        if self.is_preprocess is True:
            X = preprocess(X)
            
        res_vector = numpy.zeros(shape=X.shape[0])
        for num, x in enumerate(X):
            delta = x - self.X
            #计算距离向量
            dist_vector = numpy.sum(numpy.sqrt(delta ** 2), axis=1)
            
            id_list = numpy.argsort(dist_vector)[0: self.k] # 索引顺序按照距离由小到大排列
            d = {}
            for idx in id_list:
                key = self.y[idx]
                d[key] = d.get(key, 0) + 1
            #sorted()的返回值d是一个list
            d = sorted(d.iteritems(), key=operator.itemgetter(1), reverse=True)
            res_vector[num] = d[0][0]
        if y is None:
            return res_vector
        else:
            error_rate, error_num = self.error_rate(res_vector=res_vector, y=y)
            return res_vector, error_rate, error_num
    
    
    def error_rate(self, res_vector, y):
        
        assert isinstance(y, numpy.ndarray) and y.ndim == 1
        error = numpy.sum(res_vector != y)
        error_rate = numpy.cast['float64'](error) / y.shape[0]
        
        return error_rate, error

if __name__ == '__main__':
    
    num_samples = 1000
    num_features = 3
    train_ratio = 0.5
    os.chdir('/home/zanghu/ProgramFiles/machinelearninginaction/Ch02/')
    X = numpy.zeros(shape=(num_samples, num_features), dtype='float64')
    y = numpy.zeros(shape=num_samples, dtype='float64')
    f = open('datingTestSet2.txt')
    cnt = 0
    for line in f:
        ll = line.strip('\n').split('\t')
        assert len(ll) == 4
        X[cnt] = [eval(elem) for elem in ll[0: 3]]
        y[cnt] = numpy.array([eval(ll[3])])
        cnt += 1
        if cnt >= num_samples:
            break
    X = X[: cnt]
    y = y[: cnt]
    
    #将数据集切为两半，做一次分类=============================================================
    cut_num = int(num_samples * train_ratio)
    knn = KNN_deprecated(X=X[: cut_num, :], y=y[: cut_num], k=3)
    res_vector, error_rate, error_num = knn.classify(X=X[cut_num:, :], y=y[cut_num:])
    print error_rate, error_num
    
    #10交叉验证，做十次分类取平均值===========================================================
    batch_size = num_samples * 0.1
    #将全部样本集和类标等分为10部分，每个部分作为一个样本子集，保存在列表中
    X_parts = numpy.array([X[i*batch_size: (i+1)*batch_size] for i in xrange(10)])
    y_parts = numpy.array([y[i*batch_size: (i+1)*batch_size] for i in xrange(10)])
    duplicated_index = numpy.tile(numpy.arange(10), 2)
    res_record = []
    for i in xrange(10):
        #print 'i=%d' % i
        #print 'low=%d, high=%d' % (i, i+train_ratio*10)
        idx_train = duplicated_index[i: i+train_ratio*10]
        idx_test = list(set(duplicated_index) - set(idx_train))
        X_train = numpy.concatenate(tuple(X_parts[idx_train]), axis=0)
        y_train = numpy.concatenate(tuple(y_parts[idx_train]), axis=0)
        X_test = numpy.concatenate(tuple(X_parts[idx_test]), axis=0)
        y_test = numpy.concatenate(tuple(y_parts[idx_test]), axis=0)
        #print type(X_train), type(y_train), type(X_test), type(y_test)
        #print X_train.ndim, y_train.ndim, X_test.ndim, y_test.ndim
        
        knn = KNN_deprecated(X=X_train, y=y_train, k=3)
        res_vector, error_rate, error_num = knn.classify(X=X_test, y=y_test)
        res_record.append([error_rate, error_num])
    for num, i in enumerate(res_record):
        print 'epoch: %d  error_rate: %f  error_num: %d' % (num, i[0], i[1])
    print numpy.mean(numpy.array(res_record), axis=0)
    
    #进行10交叉验证, 取平均值，固定为0.9训练+0.1测试，不支持设置train-ratio===========================
    res_record = []
    for i in xrange(10):
        X_train = numpy.concatenate([X[:i*batch_size, :], X[(i+1)*batch_size:, :]], axis=0)
        y_train = numpy.concatenate([y[:i*batch_size], y[(i+1)*batch_size:]], axis=0)
        X_test = X[i*batch_size: (i+1)*batch_size]
        y_test = y[i*batch_size: (i+1)*batch_size]
        
        knn = KNN_deprecated(X=X_train, y=y_train, k=3)
        res_vector, error_rate, error_num = knn.classify(X=X_test, y=y_test)
        res_record.append([error_rate, error_num])
    for num, i in enumerate(res_record):
        print 'epoch: %d  error_rate: %f  error_num: %d' % (num, i[0], i[1])
    print numpy.mean(numpy.array(res_record), axis=0)
```

```python
"""
机器学习算法编程练习
自定义knn算法，假设输入是design matrix，类标是一维ndarray,不使用one_hot
测试数据集采用书中配套代码中的数据集和mnist
"""
import numpy
import theano
import theano.tensor as T
import os
import operator

class KNN(object):
    
    def __init__(self, k, X, y, is_preprocess=True):

        assert isinstance(X, numpy.ndarray) and X.ndim == 2
        assert isinstance(y, numpy.ndarray) and y.ndim == 1
        self.k = k
        self.is_preprocess = is_preprocess
        if self.is_preprocess is True:
            X = self.preprocess(X)
        self.X = X
        self.y = y
        self.num_sample = y.shape[0]
        #print 'num_samples:', self.num_sample 
        
    def preprocess(self, X):
    
        assert isinstance(X, numpy.ndarray) and X.ndim == 2
        if not hasattr(self, 'feature_mean'):
            self.feature_mean = X.mean(axis=0)
        if not hasattr(self, 'std'):
            self.std = X.std(axis=0)
    
        return (X - self.feature_mean) / self.std
   
    def classify(self, X, y=None):
        
        assert isinstance(X, numpy.ndarray)
        if X.ndim == 1:
            X = numpy.array([X], dtype=X.dtype)
        assert X.ndim == 2
        if self.is_preprocess is True:
            X = self.preprocess(X)
            
        res_vector = numpy.zeros(shape=X.shape[0])
        for num, x in enumerate(X):
            delta = x - self.X
            #计算距离向量
            dist_vector = numpy.sum(numpy.sqrt(delta ** 2), axis=1)
            id_list = numpy.argsort(dist_vector)[0: self.k] # 索引顺序按照距离由小到大排列
            d = {}
            for idx in id_list:
                key = self.y[idx]
                d[key] = d.get(key, 0) + 1
            #sorted()的返回值d是一个list
            d = sorted(d.iteritems(), key=operator.itemgetter(1), reverse=True)
            res_vector[num] = d[0][0]
        if y is None:
            return res_vector
        else:
            error_rate, error_num = self.error_rate(res_vector=res_vector, y=y)
            return res_vector, error_rate, error_num
    
    
    def error_rate(self, res_vector, y):
        
        assert isinstance(y, numpy.ndarray) and y.ndim == 1
        error = numpy.sum(res_vector != y)
        error_rate = numpy.cast['float64'](error) / y.shape[0]
        
        return error_rate, error

if __name__ == '__main__':
    
    num_samples = 1000
    num_features = 3
    train_ratio = 0.5
    os.chdir('/home/zanghu/ProgramFiles/machinelearninginaction/Ch02/')
    X = numpy.zeros(shape=(num_samples, num_features), dtype='float64')
    y = numpy.zeros(shape=num_samples, dtype='float64')
    f = open('datingTestSet2.txt')
    cnt = 0
    for line in f:
        ll = line.strip('\n').split('\t')
        assert len(ll) == 4
        X[cnt] = [eval(elem) for elem in ll[0: 3]]
        y[cnt] = numpy.array([eval(ll[3])])
        cnt += 1
        if cnt >= num_samples:
            break
    X = X[: cnt]
    y = y[: cnt]
    
    #将数据集切为两半，做一次分类=============================================================
    cut_num = int(num_samples * train_ratio)
    knn = KNN(X=X[: cut_num, :], y=y[: cut_num], k=3)
    res_vector, error_rate, error_num = knn.classify(X=X[cut_num:, :], y=y[cut_num:])
    print error_rate, error_num
    
    #10交叉验证，做十次分类取平均值===========================================================
    batch_size = num_samples * 0.1
    #将全部样本集和类标等分为10部分，每个部分作为一个样本子集，保存在列表中
    X_parts = numpy.array([X[i*batch_size: (i+1)*batch_size] for i in xrange(10)])
    y_parts = numpy.array([y[i*batch_size: (i+1)*batch_size] for i in xrange(10)])
    duplicated_index = numpy.tile(numpy.arange(10), 2)
    res_record = []
    for i in xrange(10):
        #print 'i=%d' % i
        #print 'low=%d, high=%d' % (i, i+train_ratio*10)
        idx_train = duplicated_index[i: i+train_ratio*10]
        idx_test = list(set(duplicated_index) - set(idx_train))
        X_train = numpy.concatenate(tuple(X_parts[idx_train]), axis=0)
        y_train = numpy.concatenate(tuple(y_parts[idx_train]), axis=0)
        X_test = numpy.concatenate(tuple(X_parts[idx_test]), axis=0)
        y_test = numpy.concatenate(tuple(y_parts[idx_test]), axis=0)
        #print type(X_train), type(y_train), type(X_test), type(y_test)
        #print X_train.ndim, y_train.ndim, X_test.ndim, y_test.ndim
        
        knn = KNN(X=X_train, y=y_train, k=3)
        res_vector, error_rate, error_num = knn.classify(X=X_test, y=y_test)
        res_record.append([error_rate, error_num])
    for num, i in enumerate(res_record):
        print 'epoch: %d  error_rate: %f  error_num: %d' % (num, i[0], i[1])
    print numpy.mean(numpy.array(res_record), axis=0)
    
    #进行10交叉验证, 取平均值，固定为0.9训练+0.1测试，不支持设置train-ratio===========================
    res_record = []
    for i in xrange(10):
        X_train = numpy.concatenate([X[:i*batch_size, :], X[(i+1)*batch_size:, :]], axis=0)
        y_train = numpy.concatenate([y[:i*batch_size], y[(i+1)*batch_size:]], axis=0)
        X_test = X[i*batch_size: (i+1)*batch_size]
        y_test = y[i*batch_size: (i+1)*batch_size]
        
        knn = KNN(X=X_train, y=y_train, k=3)
        res_vector, error_rate, error_num = knn.classify(X=X_test, y=y_test)
        res_record.append([error_rate, error_num])
    for num, i in enumerate(res_record):
        print 'epoch: %d  error_rate: %f  error_num: %d' % (num, i[0], i[1])
    print numpy.mean(numpy.array(res_record), axis=0)
    
    #全部样本集用于训练，之后测试===========================
    knn = KNN(X=X, y=y, k=10)
    res_vector, error_rate, error_num = knn.classify(X=X, y=y)
    print 'error_rate: %f  error_num: %d' % (error_rate, error_num)
```
