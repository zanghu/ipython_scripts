```python
from pylearn2.models.mlp import MLP, ConvRectifiedLinear, Softmax, Sigmoid
from pylearn2.costs.mlp import WeightDecay
from pylearn2.space import Conv2DSpace, VectorSpace, CompositeSpace
from pylearn2.training_algorithms.sgd import SGD, MomentumAdjustor
from pylearn2.costs.cost import SumOfCosts, MethodCost
from pylearn2.termination_criteria import MonitorBased
from pylearn2.train_extensions.best_params import MonitorBasedSaveBest
#from pylearn2.stl10_oh import STL10_OH
from pylearn2.train import Train
from pylearn2.datasets.mnist import MNIST

dsm_train = MNIST(which_set='train', start=0, stop=50000, one_hot=True)
dsm_valid = MNIST(which_set='train', start=50000, stop=60000, one_hot=True)
dsm_test = MNIST(which_set='test', start=0, stop=10000, one_hot=True)

input_space = Conv2DSpace(shape=[28, 28], num_channels=1)
h2 = ConvRectifiedLinear(layer_name='h2', output_channels=64, irange=.05, kernel_shape=[5, 5], pool_shape=[4, 4], pool_stride=[2, 2], max_kernel_norm=1.9365)
h3 = ConvRectifiedLinear(layer_name='h3', output_channels=64, irange=.05, kernel_shape=[5, 5], pool_shape=[4, 4], pool_stride=[2, 2], max_kernel_norm=1.9365)
h4 = Sigmoid(layer_name='h4', dim=500, sparse_init=15)
top_layer = Softmax(max_col_norm=1.9365, layer_name='y', n_classes=10, istdev=.05)

layers = [h2, h3, h4, top_layer]
mlp_model = MLP(input_space=input_space, layers=layers)

monitoring_dataset = {'train': dsm_train, 'valid': dsm_valid, 'test': dsm_test}

total_cost = SumOfCosts(costs=[MethodCost(method='cost_from_X'), WeightDecay(coeffs=[ .00005, .00005, .00005, .00005 ])])

alg = SGD(batch_size=100, learning_rate=.01, init_momentum=.5, monitoring_dataset=monitoring_dataset, cost=total_cost, 
            termination_criterion=MonitorBased(channel_name="valid_y_misclass", prop_decrease=0., N=10))

extensions = [ MonitorBasedSaveBest(channel_name='valid_y_misclass', save_path="convolutional_network_best.pkl"), 
            MomentumAdjustor(start=1, saturate=10, final_momentum=.99)]

train = Train(dataset=dsm_train, model=mlp_model, algorithm=alg, extensions=extensions)

if __name__ == '__main__':
    train.main_loop()
```
