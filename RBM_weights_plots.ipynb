```python
#softmax claasification test, only use GB-RBM as a filter-layer, tied weights during the softmax classifier training progress
#realize TransformerDataset
from pylearn2.base import Block
from pylearn2.datasets.transformer_dataset import TransformerDataset
from theano.sandbox.rng_mrg import MRG_RandomStreams
#from pylearn2.datasets.mnist import MNIST
from pylearn2.datasets.cifar10 import CIFAR10
from pylearn2.space import VectorSpace
from pylearn2.datasets.preprocessing import GlobalContrastNormalization, ZCA, Pipeline

from softmax_pylearn2 import SoftmaxClassifier

theano.config.compute_test_value = 'off'

dsc_train = CIFAR10(which_set='train', start=0, stop=40000, one_hot=True)
dsc_valid = CIFAR10(which_set='train', start=40000, stop=50000, one_hot=True)
dsc_test = CIFAR10(which_set='test', start=0, stop=10000, one_hot=True)

pipeline = Pipeline()
processor_1 = GlobalContrastNormalization(sqrt_bias=10., use_std=True)
processor_2 = ZCA()
pipeline.items.append(processor_1)
pipeline.items.append(processor_2)

dsc_train.apply_preprocessor(preprocessor=pipeline, can_fit=True)
dsc_valid.apply_preprocessor(preprocessor=pipeline, can_fit=True)
dsc_test.apply_preprocessor(preprocessor=pipeline, can_fit=True)   

#read prameters
#W, b, c = rbm_model.params

class Trans(Block):
    def __init__(self, rbm_model, mean_propup=True, seed=19900418):
        super(Trans, self).__init__()
        self.W, self.c, self.b = rbm_model.get_params()
        #self.W = W.T
        self.mean_propup = mean_propup
        self.rbm_model = rbm_model
        self.input_space = VectorSpace(dim=self.rbm_model.n_vis)
        self.output_space = VectorSpace(dim=self.rbm_model.n_hid)
        if mean_propup is not True:
            self.theano_rng = MRG_RandomStreams(seed=seed)
        
    def __call__(self, inputs):
        
        p_matrix = T.nnet.sigmoid(T.dot(inputs, self.W) + self.c)
        
        if self.mean_propup is True:
            return p_matrix
        else:
            return self.theano_rng.binomial(p=p_matrix, size=p_matrix.shape, dtype=p_matrix.dtype)
        
    def get_input_space(self):
        return self.input_space
        
    def set_input_space(self, space):
        self.input_space = space
    
    def get_output_space(self):
        return self.output_space
        

#make dataset

dsc_train_propup = TransformerDataset(raw=dsc_train, transformer=Trans(rbm_model=grbm_model))
dsc_valid_propup = TransformerDataset(raw=dsc_valid, transformer=Trans(rbm_model=grbm_model))
dsc_test_propup = TransformerDataset(raw=dsc_test, transformer=Trans(rbm_model=grbm_model))

#X = dsm_train_propup.get_batch_design(batch_size=60000)
#print X.shape
    
#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++=
#train softmax classifier
if __name__ == '__main__':
    
    from pylearn2.training_algorithms.sgd import SGD
    from pylearn2.train import Train
    from pylearn2.termination_criteria import MonitorBased
    from pylearn2.train_extensions.best_params import MonitorBasedSaveBest
    from pylearn2.training_algorithms.sgd import MonitorBasedLRAdjuster
    from pylearn2.training_algorithms.sgd import MomentumAdjustor
    
    #dsm = MNIST(which_set='train', one_hot=True, start=0, stop=50000)
    softmax_model = SoftmaxClassifier(num_features=400, num_classes=10)
    monitoring_dataset = {'train': dsc_train_propup, 'valid': dsc_valid_propup, 'test': dsc_test_propup}
    alg = SGD(learning_rate=0.1, cost=None, batch_size=100, init_momentum=0.1, monitoring_dataset=monitoring_dataset,
              termination_criterion=MonitorBased(channel_name='valid_mean_pred_error', N=15))
    
    # note: the save_path in extensions's parameters must not be identical to Train's save_path parameter, otherwise the overwrite problem will happen
    # distiguish the action between Train's save_freq and MonitorBasedSaveBest which in Train's extenions prameter
    train = Train(dataset=dsc_train_propup, model=softmax_model, algorithm=alg,
            extensions=[MonitorBasedSaveBest(channel_name='test_mean_pred_error', save_path='softmax_test.pkl'),
                    MonitorBasedLRAdjuster(dataset_name='valid'),
                    MomentumAdjustor(start=1, saturate=20, final_momentum=.9)],
            save_path='softmax_trainsave.pkl',
            save_freq=1)
    
    train.main_loop()
            
        
```
