```python
"""
知识1：主要考察同一图像的不同channel图的对齐方式
matlab程序中使用circshift函数实现矩阵沿轴方向或反向单向置换，numpy的类似功能函数维numpy.roll.
知识2：不关是灰度图还是RGB图像，其读入后的ndarray的数据类型都是'uint8'，并没有真正意义上的的连续数值图
这是因为每张feature map的每个像素取值都是在0-255之间的整数，因此正好可以用8位二进制数表示. 
知识3：处理RGB图像时，最好先把矩阵的数据类型转为float，以免出现数值过大产生inf（比如本例），在对计算结果
绘图之前，必须把数据类型重新转换回'uint8'，这样才能保证绘图正确。
"""
import pylab
import numpy

def my_align(img_1, img_2, irange=10):
    
    assert isinstance(img_1, numpy.ndarray) and isinstance(img_2, numpy.ndarray)
    assert img_1.shape == img_2.shape
    rstop = img_1.shape[0]
    cstop = img_1.shape[1]
    
    output = img_1
    min_offset = numpy.inf
    for i in xrange(-irange, irange):
        for j in xrange(-irange, irange):
            temp = numpy.roll(numpy.roll(img_1, shift=i, axis=0), shift=j, axis=1)
            offset = numpy.sum(numpy.sqrt((img_2 - temp) ** 2))
            #print 'offset=', offset
            if offset < min_offset:
                output = temp
                min_offset = offset
                #print 'i=%d, j=%d, min_offset=%f' % (i, j, min_offset)
            
    return output

def gray2rgb(file_name, irange=10):

    f = open(file_name)
    orin = pylab.imread(f)
    f.close()
    orin = numpy.array(orin, dtype='float64')

    assert isinstance(orin, numpy.ndarray)
    height = orin.shape[0] // 3
    B = orin[:height, :]
    G = orin[height: 2*height, :]
    R = orin[2*height: 3*height, :]

    R = my_align(R, G, irange=10).reshape(height, orin.shape[1], 1)
    B = my_align(B, G, irange=10).reshape(height, orin.shape[1], 1)
    G = G.reshape(height, orin.shape[1], 1)
    
    rgb = numpy.array(numpy.concatenate([R, G, B], axis=2), dtype='uint8')
    #pylab.imshow(rgb)
    #pylab.show()
    
    return rgb

if __name__ == '__main__':
    
    path = '/home/zanghu/ProgramFiles/fergus_CV_course/assignment1/part1_'
    for i in xrange(1, 7):
        print 'epoch %d...' % i
        file_name = path + str(i) + '.jpg'
        pic = gray2rgb(file_name=file_name, irange=10)
        pylab.subplot(2, 3, i); pylab.axis('off'); pylab.imshow(pic)
    pylab.savefig('/home/zanghu/part1.png', dpi=320)
```

```python
import numpy
import pylab

def gradient(f, p):
    """
    p: 求梯度的点的坐标
    """
    assert f.ndim == 2
    x, y = p

    fx = f[x+1, y] - f[x, y]
    fy = f[x, y+1] - f[x, y]
    
    return [fx, fy]

def gradient2(f, p):
    """求p点处的二阶偏导数"""
    assert f.ndim == 2
    x, y = p
    
    f2xx = f[x+1, y] - 2*f[x, y] + f[x-1, y]
    f2yy = f[x, y+1] - 2*f[x, y] + f[x, y-1]
    f2xy = f[x+1, y+1] + f[x, y] - f[x, y+1] - f[x+1, y]
    
    return [f2xx, f2yy, f2xy]

def get_g_pic(fig_path):
    
    f = open(fig_path)
    fig = pylab.imread(f)
    f.close()
    
    g_pic = numpy.zeros(shape=fig.shape, dtype='float64')
    fig = numpy.array(fig, dtype='float64')
    for c in xrange(fig.shape[2]):
        for i in xrange(1, fig.shape[0]-1):
            for j in xrange(1, fig.shape[1]-1):
                gx, gy = gradient(fig[:, :, c], [i, j])
                g_pic[i, j, c] = numpy.sqrt(gx**2 + gy**2)

    g_pic = numpy.array(g_pic, dtype='uint8')
    return g_pic
    
    

kn = get_g_pic('/home/zanghu/kn.jpg')
pylab.subplot(1, 2, 1); pylab.axis('off'); pylab.imshow(kn)
flower = get_g_pic('/home/zanghu/flower1.png')
pylab.subplot(1, 2, 2); pylab.axis('off'); pylab.imshow(flower)
pylab.savefig('/home/zanghu/g_pic.png')
pylab.show()
```
